```python
# This is a first exploration of my techno playlist
# Scrapped on 25/09/19 via webapp
# pafitis

import scipy.stats as stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

### Data processing and cleaning

```python
df = pd.read_csv('data/thankyoufornotsmoking.csv', sep=",")
```

```python
# does it work?
# df.head()
df
df.shape
```

```python
# first observation is that the date column is inconsistent
# lets ignore days and months and extract only the year
df['year'] = df['release'].str[0:4]

# make sure its numeric
df['year'] = df['year'].apply(pd.to_numeric)
df['year'].dtype

# we no longer need the release column
df = df.drop('release', 1) # 1 represents the columns (0 for index)

```

```python
# time to convert length to something useful
# lets use seconds
time = pd.DatetimeIndex(df['length'])

# time is now a DatetimeIndex object(?) with information on
# year (all the same / systime), 
# hours (which in reality is minutes),
# minutes (which are seconds)

time[1:7]

# lets fix this to our format of choice
seconds = time.hour*60 + time.minute

# seconds is now no. of seconds for each song
# add it to df
df['seconds'] = seconds


# drop length column
df = df.drop('length', 1)
```

```python
# save the cleaned file
df.to_csv('data/tfns-cleaned.csv')

df
```

```python
from sklearn.decomposition import PCA
pca_unscaled = PCA(n_components=2)
pca_unscaled.fit(numericdf)
```

```python
print(pca_unscaled.explained_variance_ratio_)
```

Now that I think about it, rescaling the data could affect our results. It might be better, given we are working with PCA an approach dependent on scaling/ projecting it could be worth doing this. Its fast anyhow.

The important thing to take from here is that each feature is measured in various scales. Rescaling will stop features from dominating our analysis

```python
# setup scaler
from sklearn.preprocessing import StandardScaler
scl = StandardScaler()

scl.fit(numericdf).mean_
# scl.fit(numericdf)
```

```python
numericdf.mean() #sanity check
```

```python
# obtain scaled design matrix
scaleddf = scl.transform(numericdf)
# sanity cheeeeeeeck!
len(scaleddf)
pd.DataFrame(scaleddf).std() # should have unit variance
pd.DataFrame(scaleddf).mean() # should have zero mean
```

```python
# initiate PCA again
pcascaled = PCA(n_components=2)
Xproject = pcascaled.fit(scaleddf)

Xproject.explained_variance_ratio_
```

This is too low!

- 23% for PC1
- 15% for PC2

compared to 

- 93% for PC1
- 3% for PC2

What does this say? Unscaled data could drastically affect a principal component, as is the case. One variable (or more) could have big impact on a PC since their variance range is big and dominating over the others

```python
# important to calculate loading matrix
# see here
# https://stats.stackexchange.com/questions/143905/loadings-vs-eigenvectors-in-pca-when-to-use-one-or-another

loadMatrix = pcascaled.components_.T * np.sqrt(pcascaled.explained_variance_)
loadMatrix = pd.DataFrame(loadMatrix)
loadMatrix.columns=("PC1", "PC2")
# rows represent
# "bpm", "energy", "dance", "loud", "valence", "acoustic", "pop", "year", "seconds")
```

```python
loadMatrix
```

```python
# collect transformation/ projection data to be used to plot
# each data point in PC axes
pca_data_scaled = pcascaled.transform(scaleddf)
```

```python
# Scree plot
# explained variance per PC
per_var = np.round(pcascaled.explained_variance_ratio_*100, decimals = 1)
labels = ["PC" + str(x) for x in range(1, len(per_var)+1)]

plt.bar(x=range(1, len(per_var)+1),
       height=per_var,
       tick_label = labels)
plt.ylabel("Percentage of explained variance")
plt.xlabel("Principal Component")
plt.title("Scree Plot")
plt.show()
```

```python
indexes = ["bpm", "energy", "dance", "loud", "valence", "acoustic", "pop", "year", "seconds"]
pca_df = pd.DataFrame(pca_data_scaled, 
                      columns = labels)

plt.scatter(pca_df.PC1, pca_df.PC2)
plt.title("PCA graph")
plt.xlabel("PC1")
plt.ylabel("PC2")
```

There isn't a clear seperation between data points. That's okay given how similar each song is in my playlist.

To see which variables affected our PCA

```python
loadingScores = pd.Series(pcascaled.components_[0], index = indexes)
sortedLoadingScores = loadingScores.abs().sort_values(ascending = False)

print(sortedLoadingScores)
```

Lets extend this and build a quick support vector machine (SVM)

```python
from sklearn import svm
clf = svm.OneClassSVM(kernel = "rbf",
                     gamma = 0.001,
                     nu = 0.3)

clf.fit(pca_data_scaled) # use transformation data from our scaled PCA
```

```python
xx, yy = np.meshgrid(np.linspace(-200, 500, 200),
                    np.linspace(-200, 500, 200)) # build a square space for x,y

Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]) # important that we chose 2 Principal components
                                                         # helps with visualising
Z = Z.reshape(xx.shape)

plt.figure(1, figsize=(30,20))
plt.contourf(xx, yy, Z, levels = np.linspace(Z.min(), 0, 7),
            cmap=plt.cm.Blues_r)
a = plt.contour(xx, yy, Z, levels = [0], linewidths = 5, colors = "red")
plt.contourf(xx, yy, Z, levels = [0, Z.max()], colors = "orange")







b1 = plt.scatter(pca_data_scaled[:, 0], pca_data_scaled[:, 1], 
                 c = "white", s = 100) # s is for size of point
plt.axis("tight")
plt.xlim((-100,100))
plt.ylim((-100,100)) # initial scale was too big

leg = plt.legend([a.collections[0], b1],
                ["learned frontier", "training observations"],
                loc = "upper left", prop= {"size":30}, frameon = False) # Python has weird syntax on prop size

plt.show()
```

Not enlightening! 

Is it because of the similarity between songs? 

There are 153 songs in the playlist... they can't all be the same!

### Inference

Blue represents music from my personal playlist. 
Orange represents music coming from the "Techno Bunker" playlist

Let's first start from the 

#### Similarities:

- Year of Release: 
    - This makes sense, both from my personal taste as well as the kind of techno artist that are available on Spotify. Many old school artists are not available on the platform (not all but a few) hence you end up listening to mostly newly released songs (past 5 years)
- Seconds:
    - Again, most techno songs follow the same pattern. Long build-up, breakdown, build-up, end. This usually runs around 6-8 minutes. This confirms the two bell shapes. Notice mine (blue) has a long leg indicating some songs run for more than 13 minutes. Hey Ricardo Villalobos!
- Acoustic-ness:
    - Pretty self-explanatory. This is an analysis on techno playlists!


#### Differences:
- BPM:
    - Although there's a big overlap between the two shapes, it seems that I prefer slower songs with the exception of sometimes going crazy, with tracks having 170-190 BPMs. There's greater dispersion in my tracks indicating that my playlist features tracks from wider backgrounds (chicago, london, ...)
- Energy, Loudness, Dancability:
    - Especially interesting here.
    - Although my songs range to higher BPM values, my tracks are ranked with lower Energy metrics. Might be counterintuitive at first but this can be a consequence of the subjective measurement of "energy". Some could consider "darker" or repeatitive sounds to be less energetic? We can't know exactly unless there is a specification on how Spotify ranks that. 
    - Loudness follows the same pattern. I can see how I rank lower though. I prefer more monotonous or "minimal" sounds which usually aren't the loudest. When I think loud, Amelie Lens comes to mind. That's not what is typically in my playlist (although she is good).
    - Here we see a complete switch! Even if I ranked low on energy and loudness, it seems that my songs are dance-indusers! Partyyyy.
- Valence:
    - A measurement of "happyness". Don't know if we should be taking it into consideration given how ambiguous any "happyness" measurements are in the first place but more specifically when considering techno songs. Regardless, if we do so then I'm happier than your average techno-head. Take that.
- Popularity:
    - Not to brag or anything... but the graph speaks for itself.
